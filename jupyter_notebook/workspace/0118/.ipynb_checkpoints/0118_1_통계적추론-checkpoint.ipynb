### 참고. 정규분포 관련 함수들
- stat.norm(평균,표준편차) : 평균과 표준편차의 정규분포를 따르는 확률변수 생성
- .pdf(확률변수) : 밀도함수 계산
- .cdf(확률변수) : 누적분포함수 P(X<=확률변수)
- .isf(alpha) : 상위 alpha%가 되는 확률변수값 반환
- .interval(alpha) : 확률이 alpha가 되는 가운데 부분의 구간 반환

## 추정(estimation)

### 모평균의 구간 추정

**문제 1**
표준정규분포로부터 뽑아 95% 신뢰구간을 구하는 것을 20번 반복했을 때, 몇 개의 신뢰구간이 모평균을 포함할 지 확인

```python
# 성적 데이터

df = pd.read_csv('./data/ch4_scores400.csv')
scores = np.array(df['score'])
```

```python
# 성적데이터 모집단의 평균과 분산 계산
p_mean = np.mean(scores)
p_var = np.var(scores)

p_mean
p_var
```

```python
# 모집단의 분포
xs = np.arange(101)
rv = stats.norm(p_mean, np.sqrt(p_var))
plt.plot(xs, rv.pdf(xs), color='grey')
plt.hist(scores, bins=100, range=(0,100), density=True)
plt.show()
```

```python
# sample 데이터 생성
n_sam = 10000
n = 20
samples = np.random.choice(scores, (n_sam,n)) # 20개의 값이 들어있는 리스트를 10000개 생성
samples[:5]
```

```python
rv = stats.norm() # 정규분포 객체 생성
rv.isf(0.025) # 0.025가 되는 값을 구해라
```

```python
# 모분산을 아는 경우 95% 신뢰구간 : 정규분포를 이용
rv = stats.norm() # 정규분포 객체 생성
n_samples = 20 # 표본의 크기

plt.vlines(p_mean,0, 21) # 평균값으로 수직선그리기
for i in range(n_samples):
    sample_ = samples[i]
    s_mean = np.mean(sample_)
    lcl = s_mean - rv.isf(0.025) * np.sqrt(p_var/n) # 모분산/n
    ucl = s_mean + rv.isf(0.025) * np.sqrt(p_var/n) # ucl = s_mean - rv.isf(0.975) * np.sqrt(n_var/n)
    
    if lcl <= p_mean <= ucl :
        plt.scatter(s_mean,n_samples-i,color='grey')
        plt.hlines(n_samples-i, lcl, ucl, color='grey')
    else :
        plt.scatter(s_mean,n_samples-i,color='b')
        plt.hlines(n_samples-i, lcl, ucl, color='b')
        
plt.xticks([p_mean])
plt.xlabel('popualtion mean')
plt.show()
```

```python
# 모분산을 모르는 경우 95% 신뢰구간 : t-분포를 이용
n=20
alpha = 0.025
sample_ = samples[0]
s_mean = np.mean(sample_)
s_var = np.var(sample_)
rv = stats.t(df=n-1)
lcl = s_mean - rv.isf(alpha) * np.sqrt(s_var/n)
ucl = s_mean + rv.isf(alpha) * np.sqrt(s_var/n)
lcl,ucl
```

```python
n_samples = 20 # 표본의 크기
n = 20
rv = stats.t(df=n-1)
alpha = 0.025

plt.vlines(p_mean,0, 21) # 평균값으로 수직선그리기
for i in range(n_samples):
    sample_ = samples[i]
    s_mean = np.mean(sample_)
    s_var = np.var(sample_)
    lcl = s_mean - rv.isf(alpha) * np.sqrt(p_var/n) # 모분산/n
    ucl = s_mean + rv.isf(alpha) * np.sqrt(p_var/n) # ucl = s_mean - rv.isf(0.975) * np.sqrt(n_var/n)
    
    if lcl <= p_mean <= ucl :
        plt.scatter(s_mean,n_samples-i,color='grey')
        plt.hlines(n_samples-i, lcl, ucl, color='grey')
    else :
        plt.scatter(s_mean,n_samples-i,color='b')
        plt.hlines(n_samples-i, lcl, ucl, color='b')
        
plt.xticks([p_mean])
plt.xlabel('popualtion mean')
plt.show()
```

## 통계적 가설검정(test)

대립가설(H1) : µ < 130

```python
import pandas as pd
```

```python
# 감자 무게 측정 데이터
df = pd.read_csv('./data/ch11_potato.csv')
sample = np.array(df['무게'])
sample
```

```python
# 표본 분산 확인
s_var = np.var(sample)
s_var
```

```python
# 구간 추정 , 모분산 = 9
rv = stats.norm(130,np.sqrt(9/14))
rv.isf(0.95)
```

```python
# 검정 통계량
Z = (s_mean - 130)/np.sqrt(9/14)
Z
```

```python
rv = stats.norm()
rv.isf(0.95)
```

## 가설검정에서의 두 가지 오류
- 제1종 오류
    - 실제로 평균이 130g인데도 평균은 130g보다 작다라는 결론을 내리는 상황
    - 본래 검출하지 말아야 할 것을 검출한 것이므로 오탐(false negative)
- 제2종 오류

```python
# 모분산을 아는 경우 모평균에 대한 양측검정을 위한 함수
# 정규분포 사용

def pmean_test(sample, mean0, p_var, alpha=0.05):
    s_mean = np.mean(sample)
    n = len(sample)
    rv = stats.norm()
    interval = rv.interval(1-alpha)

    z = (s_mean - mean0) / np.sqrt(p_var/n)
    if interval[0] <= z <= interval[1]:
        print('귀무가설을 채택')
    else:
        print('귀무가설을 기각')

    if z < 0:
        p = rv.cdf(z) * 2
    else:
        p = (1 - rv.cdf(z)) * 2
    print(f'p값은 {p:.3f}')
```

```python
# 감자 튀김의 표본 데이터로 확인
pmean_test(sample,130,9)
```

```python
# 양측검정, t-분포
def pmean_test(sample, mean0, alpha=0.05):
    s_mean = np.mean(sample)
    u_var = np.var(sample, ddof=1)
    n = len(sample)
    rv = stats.t(df=n-1)
    interval = rv.interval(1-alpha) # 임계값 계산 

    t = (s_mean - mean0) / np.sqrt(u_var/n)
    if interval[0] <= t <= interval[1]:
        print('귀무가설을 채택')
    else:
        print('귀무가설을 기각')

    if t < 0:
        p = rv.cdf(t) * 2
    else:
        p = (1 - rv.cdf(t)) * 2
    print(f'p값은 {p:.3f}')
```
