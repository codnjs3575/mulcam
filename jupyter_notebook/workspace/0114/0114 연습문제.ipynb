### 연습문제1.
- 두 개의 데이터프레임을 만들고 merge 명령을 병합한다.
- 단, 데이터 프레임은 다음 조건을 만족해야 한다.
    1. 각각 5X5 이상의 크기를 가진다.
    2. 공통열을 하나이상 갖는다. 
        - 다만 공통열의 이름은 서로 다르게 할 것
    3. merge의 경우를 inner, outer, left, right 4개의 형태로 출력할 것
    4. 지정된 인덱스와 컬럼명을 갖는다. 

```python
data = [[22, 60.1, 170.5, '남', '서울'], 
        [45, 51.3, 157.3, '여', '부산'], 
        [22, 68.3,180.1,  '남', '대구'],
        [33, 88.3, 190.2, '남', '제주'], 
        [27, 48.3, 160.1, '여', '강릉']]

df1 = pd.DataFrame(data, 
                   index=['홍길동', '이몽룡', '성춘향','변학도','김연아'], 
                   columns=["나이", "몸무게", "키", "성별" ,"주소"])
df1
```

```python
data = [[22, 60.1, 170.5, '남', '서울'], 
        [45, 51.3, 157.3, '여', '부산'], 
        [22, 68.3,180.1,  '남', '울산'],
        [33, 88.3, 190.2, '남', '제주'], 
        [27, 48.3, 160.1, '여', '광주']]
df2 = pd.DataFrame(data,
                   index=[1,2,3,4,5],
                   columns=["나이", "몸무게", "키", "성별" ,"지역"])
df2
```

```python
df1.head(3)
df2.head(3)
```

```python
pd.merge(df1,df2,left_on = '주소',right_on = '지역')
pd.merge(df1,df2,how='outer',left_on = '주소',right_on = '지역')
pd.merge(df1,df2,how='left',left_on = '주소',right_on = '지역')
pd.merge(df1,df2,how='right',left_on = '주소',right_on = '지역')
```

### 연습문제2. 
- 어느회사의 전반기(1월-6월)실적을 나타내는 데이터프레임과
후반기(7월-12월)실적을 나타내는 데이터 프레임을 작성한 후
합친다(단순병합)
- 실적 정보는 "매출","비용","이익"으로 이루어진다.
- 이익=매출-비용 : 가공필드로 생성
    
- 또한 1년간의 총 실적을 마지막 열로 덧붙인다.

**상반기(1-6월) 실적 데이터프레임 생성**
- month 변수 생성

```python
month = [str(i)+'월' for i in range(1,13)]
month
```

- 1~6월의 매출, 비용 데이터 생성(딕셔너리)

```python
# 매출, 비용 data random  생성
np.random.seed(3)
num = []
for i in range(12):
    num.append(np.random.randint(1,100,2))

data = {}
for i in range(6):
    data[month[i]] = num[i]
data
```

- 1~6월의 매출과 비용 데이터프레임 생성

```python
# 데이터프레임 생성
df1 = pd.DataFrame([i for i in data.values()],
    index = [i for i in data.keys()],
    columns = ['매출','비용']
        )
df1 = df1.T
df1
```

- '이익' 행이 추가된 데이터 프레임 작성

```python
# '이익' 행 생성 : 매출 - 비용
df1.loc['이익'] = df1.loc["매출"] - df1.loc["비용"]
df1
```

**하반기(7~12월) 실적 데이터프레임 생성**
- 매출, 비용으로 구성

```python
# 매출, 비용 data random  생성
np.random.seed(3)
data = {}
for i in range(6,12):
    data[month[i]] = num[i]
data

# 데이터프레임 생성
df2 = pd.DataFrame([i for i in data.values()],
    index = [i for i in data.keys()],
    columns = ['매출','비용']
        )
df2 = df2.T
df2

# 가공 행 이익 생성
df2.loc['이익'] = df2.loc["매출"] - df2.loc["비용"]
df2
```

**상반기 실적 데이터프레임과 하반기 실적 데이터프레임 연결**

```python
df3 = pd.merge(df1,df2,left_index=True,right_index=True)
df3
```

- '총실적' 열 추가

```python
df3['총실적'] = df3.sum(axis=1)
df3
```

## 그룹함수  및 피봇 테이블 이용 간단한 분석 예제

#### 식당에서 식사 후 내는 팁(tip)과 관련된 데이터 : tips

- seaborn 패키지 내 tips 데이터셋 사용

    - total_bill: 식사대금

    - tip: 팁

    - sex: 성별

    - smoker: 흡연/금연 여부

    - day: 요일

    - time: 시간

    - size: 인원

```python
tips = sns.load_dataset("tips")
tips.head(3)
```

##### 식사 대금 대비 팁의 비율이 언제 가장 높아지는가?

- 가공 필드 생성 : 식사대금 대비 팁의 비율
    - tip_pt = 팁 / 식사대금

```python
tips["tip_pt"] = tips["tip"]/tips["total_bill"]
tips
```

```python
tips.info()
```

```python
tips.describe()
```

```python
# 성별 인원수를 계산
tips_g = tips.groupby('sex')
tips_g.count()
tips_g.size()
```

```python
# 흡연 유무에 따른 성별 인원
a = tips.value_counts(['sex','smoker'])
a
```

```python
# 흡연 유무에 따른 성별 인원을 피봇테이블로 구현
a.unstack()
tips.pivot_table(index='sex',columns='smoker',values='tip',aggfunc="count")
```

```python
# 성별 팁 비율의 평균
tips.groupby('sex')['tip_pt'].mean()
```

```python
# 흡연 유무에 따른 팁 비율의 평균
tips.groupby('smoker')['tip_pt'].mean()
```

```python
# 성별과 흡연 유무에 따른 팁 비율의 평균

tips.pivot_table(index='sex',columns='smoker',values='tip_pt',aggfunc="mean")
```

##### 여성 혹은 흡연자의 팁 비율이 조금 높고, 여성 흡연자가 팁을 많이 줌 

```python
# 평균 통계량만 확인 했으므로 다른 통계값도 확인
tips.groupby(['sex','smoker'])['tip_pt'].describe()
```
