### 연습문제 

1. 타이타닉 데이터에서 승객의 성별(sex) 인원수, 나이별(age) 인원수, 선실별(class) 인원수, 사망/생존(alive)인원수를 구하시오.

2. 성별 인원수는 인덱스 기준으로 정렬하고, 나이별 인원수는 값 기준으로 정렬하며, 나머지는 임의 기준으로 선정하여 정렬 하시오.

```python
import seaborn as sns # seaborn 패키지 : 그래프 패키지

titanic = sns.load_dataset('titanic')
titanic.head() 
```

```python
# 타이타닉 승객에 대하여 성별 인원수 구하기
titanic['sex'].value_counts().sort_index()
```

```python
# 타이타닉 승객에 대하여 나이별 인원수 구하기
titanic['age'].value_counts().sort_values()
```

```python
# 타이타닉 승객에 대하여 선실별 인원수 구하기
#파이썬에서는 형태 분류 class로 작업

titanic['class'].value_counts().sort_values(ascending=False)
```

```python
#타이타닉 승객에 대하여 사망/생존 인원수 구하기

titanic['alive'].value_counts()
```

```python
#타이타닉 승객에 대하여 사망/생존 인원수 구하기
#.연산자 이용해서 data에 접근

titanic.alive.value_counts()
```

### 연습문제 : 수치형-> 범주형 데이터 생성

: 타이타닉호 승객을 사망자와 생존자 그룹으로 나누고(alive), 각 그룹에 대해 미성년자, 청년, 중년, 장년, 노년 승객의 비율을 구하시오.

- bins=[1,15,25,35,60,99]
- labels=['미성년자','청년','중년','장년','노년']
- 각 그룹별 비유의 전체 합은 1이 되어야 한다.

**타이타닉 데이터셋 로딩**

```python
# 타이타닉 승객 dataset 읽어오기
import seaborn as sns 
titanic = sns.load_dataset('titanic')
titanic.head()
titanic.info()
titanic.count()
```

**데이터 셋 분리 : 생존자 그룹과 사망자 그룹**

```python
# 사망자 그룹 데이터 프레임 추출
titanic_0 = titanic[titanic["alive"]=='no']
titanic_0
```

```python
# 생존자 그룹 데이터 프레임 추출
titanic_1 = titanic[titanic["alive"]=='yes']
titanic_1
```

**데이터 범위 설정**

```python
# 라벨과 범위 간격 설정
bins=[0.1,15,25,35,60,99]
labels=['미성년자','청년','중년','장년','노년']
```

```python
# 범위 유효성 검사
# bins=[1,15,25,35,60,99]
#제시된 bins의 최초값 1 이하인 값 1.0과 0.42가 각
#그룹의 최소나이로 확인되므로 bins 수정
#최대나이는 74와 80으로 확인되어서 수정 필요 없음

min(titanic_0['age'])
max(titanic_0['age'])
min(titanic_1['age'])
max(titanic_1['age'])

```

**범주형 데이터 생성**

```python
# 카테고리 데이터 생성
titanic_0['age_cut'] = pd.cut(titanic_0.age,bins,labels=labels)
titanic_1['age_cut'] = pd.cut(titanic_1.age,bins,labels=labels)
```

**범주형 데이터 확인**

```python
#카테고리 데이터 확인
titanic_0.head()
titanic_1.head()
```

**범주형 데이터 빈도 계산**

- 사망자 그룹의 연령대별 승객수

```python
# 카테고리 데이터들의 빈도 수 계산
titanic_0['age_cut'].value_counts()

# 사망자그룹 계산
pd.value_counts(titanic_0['age_cut']).sum()

# 각 연령대별 승객수의 비율 계산
pd.value_counts(titanic_0['age_cut'], normalize = True)

# 전체 비율의 합 계산
pd.value_counts(titanic_0['age_cut'], normalize = True).sum()
```

- 생존자 그룹 빈도

```python
# 카테고리 데이터들의 빈도 수 계산
titanic_1['age_cut'].value_counts()

# 생존자그룹 계산
pd.value_counts(titanic_1['age_cut']).sum()

# 각 연령대별 승객수의 비율 계산
pd.value_counts(titanic_1['age_cut'], normalize = True)

# 전체 비율의 합 계산
pd.value_counts(titanic_1['age_cut'], normalize = True).sum()
```
